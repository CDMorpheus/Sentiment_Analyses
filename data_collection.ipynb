# Dictionary of drug name and slangs

```python
#importing files
import PyPDF2
import numpy as np
import io
import requests
from PyPDF2 import PdfReader
import re
#extracting drug names/ slangs from pdf on a website
headers = {'User-Agent': 'Mozilla/5.0 (X11; Windows; Windows x86_64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/103.0.5060.114 Safari/537.36'}

url = 'https://bhddh.ri.gov/sites/g/files/xkgbur411/files/documents/Drug-Slang---guide-FINAL.pdf'

response = requests.get(url=url, headers=headers, timeout=120)
on_fly_mem_obj = io.BytesIO(response.content)

#reading pdf
pdfReader = PdfReader(on_fly_mem_obj)

#data from first page
pageObject = pdfReader.getPage(0)
text= (pageObject.extractText())
sentences = text.splitlines()
start= sentences.index('Common Drug Slang:  ')
data= sentences[(start+2):]
```

Isolating each drug/drug slang word from page 1

```python
#processig words
new=' '.join(data)
a=(new.replace(':', ',')).replace('-','').split(',')
a=[(x.replace(' ', '-')).replace('---', ',') for x in a]
new=(','.join(a)).split(',')
ner1= ([x.strip('-') for x in a if (len(x) < 18 and len(x)>4)])
ner1 = re.split(',', ','.join(ner1))
ner1
```

Isolating each drug/drug slang word from page 2

```python
pageObject = pdfReader.getPage(1)
 
text= (pageObject.extractText())
sentences = text.splitlines()
```

```python
new=' '.join(sentences[1:41])
a=(new.replace(':', ',')).replace('-','').split(',')
a=[(x.replace(' ', '-')).replace('---', ',') for x in a]
new=(','.join(a)).split(',')
ner= ([x.strip('-') for x in a if (len(x) < 18 and len(x)>4)])
ner = re.split(',', ','.join(ner))
ner
```

```python
final_list= [x for x in np.hstack((ner1,ner)) if (')' or '(') not in x] #combining words from page1 and page2
len(final_list)
```

# Extracting dialogues from series: Breaking Bad 
mentioning the website in google search so that formatting is consistent throughout all the iterations over different episodes.

```python
from googlesearch import search
import requests
from bs4 import BeautifulSoup
import re

neg_data=[]
for season in range (1, 6):
    #google search phrase:
    query_ = "S"+str(season)+ " E"

    #iterating over episodes to gather text data:
    for episode in range (1,13):
        query= query_ + str(episode)+ " Breaking Bad- Transcripts - Forever Dreaming"
        web_result= [x for x in (search(query, num_results=3))]
        print(season, episode)
        subs='foreverdreaming.org/viewtopic.php'
        res = [x for x in web_result if re.search(subs, x)]
        if res:
            temp=0
            for i in range (3):
                print(res[i]) #web link
                response = requests.get(res[0],headers=headers)
                soup = BeautifulSoup(response.text, 'html.parser')
                text_data = soup.get_text()
                if len(text_data)<10000:
                    continue
                else:                    
                    print('here', len(text_data))
                    text_data= text_data.replace('.',' ')
                    # generate the pattern
                    pattern = r'.*\b(?='+' | '.join(final) + r')\b.*'
                    # resulting pattern .*\b(?=drug1|drug2|drug3|drug4|drug5)\b.*
                    neg_data.append('\n'.join(re.findall(pattern, text_data, re.IGNORECASE)))
                    neg_data.append('---------------------------------------------')
                    break
                
            
print(neg_data)
```

```python
#edditing the sentences
neg_sentence= '\n'.join('\n'.join(neg_data).split('---------------------------------------------'))
len(neg_sentence)
```

```python
# removing 1 repeating sentence from heading for each episode
neg= neg_sentence.replace('To provide for his family\'s future after he is diagnosed with lung cancer, a chemistry genius turned high school teacher teams up with an ex-student to cook and sell the world\'s purest crystal meth ', '')
neg
```

```python
neg1= (neg.split(' '))
```

```python
(neg1)
```

```python
pruned = []
for inp in neg1:
   if not inp.endswith(':'):
       pruned.append(inp)
pruned=(' '.join(list(filter(None, pruned)))).split('\n')
print(len(pruned))
```

```python
import pandas as pd
#saving data to xlsx file
(pd.DataFrame(pruned)).to_excel("dialogou.xlsx") 
```

# Extracting dialogues from series: Narcos

```python
from googlesearch import search
import requests
from bs4 import BeautifulSoup
import re

neg_data=[]
for season in range (1, 4):
    #google search phrase:
    query_ = "Narcos: Season "+str(season)+ " Episode "

    #iterating over episodes to gather text data:
    for episode in range (1,11):
        query= query_ + str(episode)+ " script - subslikescript"
        web_result= [x for x in (search(query, num_results=3))]
        print(season, episode)
        subs='subslikescript'
        res = [x for x in web_result if re.search(subs, x)]
        if res:
            temp=0
            for i in range (3):
                print(res[i]) #web link
                response = requests.get(res[0],headers=headers)
                soup = BeautifulSoup(response.text, 'html.parser')
                text_data = soup.get_text()
                if len(text_data)<10000:
                    continue
                else:                    
                    print('here', len(text_data))
                    text_data= text_data.replace('.',' ')
                    # generate the pattern
                    pattern = r'.*\b(?='+' | '.join(final) + r')\b.*'
                    # resulting pattern .*\b(?=drug1|drug2|drug3|drug4|drug5)\b.*
                    neg_data.append('\n'.join(re.findall(pattern, text_data, re.IGNORECASE)))
                    neg_data.append('---------------------------------------------')
                    break
                
            
print(neg_data)
```

```python
neg_sentence= ('.'.join('\n'.join(neg_data).split('---------------------------------------------'))).split('?')
(neg_sentence)
```

```python
#saving data to xlsx file
pattern = r'.*\b(?='+' | '.join(final) + r')\b.*'
# resulting pattern .*\b(?=drug1|drug2|drug3|drug4|drug5)\b.*
a=(['\n'.join(re.findall(pattern, x, re.IGNORECASE))] for x in neg_sentence)
(pd.DataFrame(a)).to_excel("dialogou_n.xlsx")  
```
